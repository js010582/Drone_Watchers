### Combine the Date & Time Columns from Weather Data into a new Column with Numerical Format.

```python
from datetime import timedelta
from datetime import datetime

pd.options.mode.chained_assignment = None  # default='warn', Turns off annoying warning

df_WD['datetime'] = '-' #Added a column called datetime to be used for the combined date time

#date_time_obj = datetime.strptime(date_time_str, '%Y-%m-%d %H:%M:%S')
#print (datetime.strptime('April 4, 2016', '%B %d, %Y'))
#print (datetime.strptime('12:41 PM', '%I:%M %p'))
#print (date_time_obj.date())
#print (date_time_obj.time())

for count in df_WD.index:
    #print (df_WD.datetime[count])
    temp_datetime = df_WD.Date[count] + ' ' + (df_WD.Time[count])
    #print(temp_datetime, end = ' ')
    #print(' Converted to ', end = ' ')
    #print(datetime.strptime(temp_datetime, '%B %d, %Y %I:%M %p'))
    df_WD.datetime[count] = (datetime.strftime(datetime.strptime(temp_datetime, '%B %d, %Y %I:%M %p'), '%Y-%m-%d %H:%M:%S'))
```
